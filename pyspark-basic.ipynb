## threshold
THRESH‚Äù Parameter in na.
In this parameter, we set the threshold value of the minimum NON NULL values in a particular row i.e. Suppose if we set the threshold value to 2, then that means the row will be dropped only if the total number of null values exceeds 2 otherwise, that row will not get dropped.

```python
df_spark2.show()
```

```python
df_spark2.na.drop(how='any',thresh=2).show()
```

```python
##Subset
df_spark2.na.drop(how="any",subset=['Name']).show()
```

# Filling the missing values 

```python
df_spark2.show()
```

```python
df_spark2.na.fill("Missing Values",['Name','age','Experience','Salary']).show()
```

```python
df_spark2.printSchema()
```

```python
df_spark2 = df_spark2.withColumn('age',df_spark2.age.cast('int'))
```

```python
df_spark2 = df_spark2.withColumn('Experience',df_spark2.Experience.cast('int'))
df_spark2 = df_spark2.withColumn('Salary',df_spark2.Salary.cast('int'))
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols = ['age','Experience','Salary'],
    outputCols = ["{}_imputed".format(c) for c in ['age','Experience','Salary']]).setStrategy("median")
```

```python
imputer.fit(df_spark2).transform(df_spark2).show()
```
